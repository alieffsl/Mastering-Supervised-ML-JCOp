```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 15, 100)
y = np.sqrt(x)
y1 = np.sin(x)
y2 = np.cos(x)
```

## Inisialisasi Figure dan Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.plot(x, y)
```

```python
fig = plt.figure(figsize = (6,6))
ax = plt.axes()
ax.plot(x, y1,'r8--', label = "Sin", linewidth = 0.5, markersize = 3)
ax.plot(x, y2,'bv-.', label = "Cos")
ax.set(xlim=(0,16), ylim=(-2, 2), title = "Sin dan Cos",xlabel="X", ylabel="Y")
# ax.set_ylim(-2, 2)
# plt.axis('equal')
ax.legend(loc='lower center', ncol=2, shadow=True, borderpad=1)
```

## Scatter Plot

```python
df = pd.read_csv('data/circle_data.csv', delimiter=';')
df.head()
```

## Subplot

### MATLAB Style

```python
plt.figure(figsize=(12,5))

plt.subplot(131)
plt.hist(numbers,bins=100,density=2,cumulative=True);

plt.subplot(132)
plt.scatter(x,y,c=rad,s=rad*100,cmap='seismic')

plt.subplot(133)
plt.hist(numbers,bins=100);
```

### Matplotlib Style

```python
fig,ax=plt.subplots(2,3,figsize=(15,10))

ax[0,0].hist(numbers,bins=100,density=2,cumulative=True);
ax[0,1].axis('off')
ax[0,2].hist(numbers,bins=100);
ax[1,0].axis('off')
ax[1,1].scatter(x,y,c=rad,s=rad*100,cmap='seismic')
ax[1,2].axis('off')
```

## Text and Annotation

```python
numbers = np.random.randn(10000)
```

```python
plt.figure(figsize=(6,6))
plt.hist(numbers,bins=90,density=1);
plt.ylim(0,0.8)
plt.axvline(0, c='g',ls='--')
plt.text(0.1,0.6,'Mean = 0', ha='left')
plt.annotate("Mean = 0", xy=(0,0.4), xytext=(1,0.5), ha='center', arrowprops={'arrowstyle': '-|>'})
```

## 3D Plot

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
```

```python
x = np.linspace(-6,6,5)
y = np.linspace(-6,6,5)

X,Y=np.meshgrid(x,y)
Z= np.sqrt(X**2 + Y**2 + 10)
```

```python
X
```

```python
Y
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,antialiased=False, cmap='winter')
ax.set(xlabel="x", ylabel="y", zlabel="z")
ax.view_init(elev=-20,azim=45)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X,Y,Z,cmap="winter")
ax.set(xlabel="x", ylabel="y", zlabel="z")
ax.view_init(elev=-30,azim=45)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.contour3D(X,Y,Z,50, cmap='winter')
ax.set(xlabel="x", ylabel="y", zlabel="z")
ax.view_init(elev=-20,azim=45)
```
