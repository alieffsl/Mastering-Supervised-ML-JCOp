- Increase n_neighbor => semakin banyak yang divoting => semakin merata => complexity decrease => prone to underfit<br>Artinya jika kita training menggunakan KNN dan overfit, maka tambahkan 'n_neighbor'

- Kernel RBF cenderung overfit<br>Kernel linear cenderung underfit
- Increase "C"/Penalty => semakin tidak toleran => semakin berliku-liku => complexity increase => prone to overfit
- increase gamma/detailness => semakin detail => semakin berliku-liku => complexity increase => prone to overfit

# Exercise - Random Forest Parameter

- increase n_estimator => more averaging => decrease overfit 
- increase max_depth => more branch => more decision => low bias => prone to overfit
- increase min_samples_split => prevent splitting => less branch => less depth => decrease overfit
- increase max_features => prone to get best decision => prone to overfit
- increase min_impurity_decrease => prevent splitting => decrease overfit
- increase min_samples_leaf => prevent splitting => less depth => decrease overfit
